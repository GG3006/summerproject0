```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
#In this beginning step, I retrieved data from 20 pages of the searching results in yellowpage using a for loop and beautiful 
#soup; everytime I scrapt the data from one page, it was cleaned a little bit and was appendded to a "data_list" 

#Tips: to find the class of the text, right click the page and inspect

#The reason I used try and pass here is that some retrieved data doesn't have a matching format, and is not useful

data_list = []
for i in range(1,20):
    r = requests.get("https://www.yellowpages.com/search?search_terms=Drug%20Abuse%20%26%20Addiction%20Centers&geo_location_terms=virginia&s=name&page="+str(i))
    soup = BeautifulSoup(r.content, 'lxml')
    bn=soup.find_all('a', attrs={'class': 'business-name'}) 
    st=soup.find_all("span", attrs={"class":"street-address"})
    ct=soup.find_all("span", attrs={"class":"locality"})

        
    for i in range(0,30):
        try:
            name = bn[i].text
            address = st[i].text
            city=ct[i].text
            city = city.replace(u'\xa0', '')
            data = [name,address,city]
            data_list.append(data)
        except:
            pass
```

```python
print(data_list)
```

```python
#Step2: write the data_list to a csv file
import csv
with open('test.csv', 'w') as file:
    a = csv.writer(file, delimiter=',')
    a.writerows(data_list)
```

```python
my_data['State']='Va'
```

```python
#Step 3: make a field of a valid geograpgic address, which could be regonized by geopy module
my_data['Full_Address']=my_data['Address']+' '+my_data['City']+my_data['State']
```

```python
#Step 4, for people who need lagitude and longitude information, we use Nominatim function here to transform address

#Specify timeout=10 here cause the default running time is only 1 minute, if your data contains 500 records, you would need
#at least 2.5 minutes

#if location is to test whether the data could be regonized by geopy, though I dont think any of data is invalid at this
#point, but it would pop up an error without the if clause.

#from geopy.exc import GeocoderTimedOut
geolocator = Nominatim()

for i in range(0,525):
    location = geolocator.geocode(my_data.iloc[i, :]['Full_Address'], timeout=10)
    if location:
        my_data.iloc[i,-2]=location.latitude
        my_data.iloc[i,-1]=location.longitude
    else :
        pass
```
